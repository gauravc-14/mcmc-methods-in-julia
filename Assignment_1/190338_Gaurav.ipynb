```julia
using Plots
using Statistics
```

```julia
container = []
lmd = 6
epoch = 20000
```

```julia
for n in 1:epoch
    i = 0
    p=exp(-lmd)
    CDF = p
    U = rand()
    while(true)
        if U<CDF
            push!(container,i)
            break
        else
            p = p*(lmd/(i+1))
            i = i+1
            CDF = CDF + p
            end
    end 
end
```

```julia
mean(container)
```

```julia
var(container)
```

```julia
histogram(container)
```
